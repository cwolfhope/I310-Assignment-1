# Analyzing the Top 100 Zac Mains for Trends in Personal Statistic Data

```python
import pandas as pd
import seaborn as sns

ZR = pd.read_csv("Zac Rankings 9-17-22.csv")
```

```python
ZR.head()
```

Imports the relevant Pandas and Seaborn libraries, reads in the rankings dataset, and builds a dataframe using the data.

```python
sns.scatterplot(x='Rank', y='Winrate', data=ZR, hue='Games Played')
```

Graph #1: Displays the winrate of the Top 100 Zac Mains as a measure of rank, in order to guage the relationship between winrate and rank. The number of games played is displayed so that correlations between rank and account age can be observed.

```python
sns.set_style('whitegrid')
sns.lmplot(x='Rank', y='Winrate', data=ZR)
```

Graph #2: Uses the same scatterplot as graph 1. Shows the linear regression data of graph 1. 

```python
ZR['KDA Ratio'] = ((ZR['Average Kills'] + ZR['Average Assists']) / ZR['Average Deaths'])
```

```python
ZR.head()
```

Adds collumn to dataset dataframe displaying the KDA Ratio ((K+A)/D) for utilization in subsequent graphs.

```python
sns.scatterplot(x='KDA Ratio', y='Winrate', data=ZR, hue='Rank')
```

Graph #3: Displays the Winrate of players with respect to KDA Ratio such that correlations can be observed how Winrates are Influenced by KDA Ratio. Rank is also displayed so that correlations between KDA Ratio and Rank can be observed.

```python
sns.scatterplot(x='KDA Ratio', y='Winrate', data=ZR, hue='Average Kills')
```

Graph #4: Uses the same scatterplot as graph 3. Shows the hue of average kills such that correlations between Average Kills and Winrate can be observed.

```python
sns.scatterplot(x='KDA Ratio', y='Winrate', data=ZR, hue='Average Assists')
```

Graph #5: Uses the same scatterplot as graph 3. Shows the hue of average assists such that correlations between Average Assists and Winrate can be observed.

```python
sns.scatterplot(x='KDA Ratio', y='Winrate', data=ZR, hue='Average Deaths')
```

Graph #6: Uses the same scatterplot as graph 3. Shows the hue of average deaths such that correlations between Average Deaths and Winrate can be observed.

```python
sns.set_style('whitegrid')
sns.lmplot(x='KDA Ratio', y='Winrate', data=ZR)
```
